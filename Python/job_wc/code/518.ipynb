```python
import requests as r
from bs4 import BeautifulSoup
import re
from collections import Counter
```

```python
lc = Counter()
count = 0
for i in range(1,55):
    try:
        Host = 'https://www.518.com.tw'
        url = Host+'/job-index-P-{}.html?i=1&am=1&ab=2032001,2032002,'.format(i)
        res = r.get(url)
        soup = BeautifulSoup(res.text,'lxml')
        rows = soup.select('li.title > a')
        
        for row in rows:
            url2 = row['href']
            print(url2)
            res2 = r.get(url2)
            soup2 = BeautifulSoup(res2.text,'lxml')
            
            content = soup2.select_one('div.job-detail-box > dl')
            text = content.text.replace('\n','')
            
            content2 = soup2.select_one('div.JobDescription > p')
            text2 = content2.text.replace('\n', '')
            
            word = re.findall('object c|visual basic|\.net|[A-Za-z]+[+#]*' , text, re.IGNORECASE)
            word2 = re.findall('object c|visual basic|\.net|[A-Za-z]+[+#]*' , text2, re.IGNORECASE)
#             print(word)
#             print(word2)
            word_count = word + word2
            print(word_count)
            pure_word = list(set(word_count))
            print(pure_word)
            for ind in pure_word:
                if ind in lc:
                    lc[ind] += 1
                else:
                    lc[ind] = 1
#             print(lc.most_common())
            count += 1
            print('COUNT = {}'.format(count))
    except Exception as e:
        print(e)
        continue
        
    print(lc.most_common())
    print('[Debug] Page {} is finished'.format(i)) 
lc.most_common()
```
