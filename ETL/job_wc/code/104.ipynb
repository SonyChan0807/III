```python
import requests as r
from bs4 import BeautifulSoup
import pandas as pd
from collections import Counter
import time

wc = Counter()
Host = 'https://www.104.com.tw'
lan_list = []

for i in range(1,300):
    try:
#         if i % 27 == 0:
#             time.sleep(10)

        url = Host + "/jobbank/joblist/joblist.cfm?jobsource=n104bank1&ro=1&keyword=%E8%BB%9F%E9%AB%94%E5%B7%A5%E7%A8%8B%E5%B8%AB&order=1&asc=0&page={}".format(i)
    #     print(url)
        res = r.get(url)
        soup = BeautifulSoup(res.text, 'lxml')
        rows = soup.select(".j_cont > ul > li  > div > a")

        for ind in rows:
            href = Host+ind['href']
            if( "jobno" in href):  
                lan_list.append(href)
                print(len(lan_list))
    #             print(href)
                res2 = r.get(href)
                soup = BeautifulSoup(res2.text, 'lxml')
                lans = soup.select('.tool')[0].select('a')

                for lan in lans :
                    if lan.text in wc:
                        wc[lan.text] +=1
                    else :
                        wc[lan.text] = 1
    except:
        continue
    print('[Debug] Page {} is finished'.format(i))
df = pd.DataFrame(wc.most_common())
lan_csv = df.to_csv('lan.csv')
df.head(10)
# 28  63  
```
